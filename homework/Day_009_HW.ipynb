```python
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

x=np.random.randint(0,50,1000)
y=np.random.randint(0,50,1000)

np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
```

```python
x=np.random.randint(0,50,1000)
y=x+np.random.randint(0,10,1000)

np.corrcoef(x,y)
```

```python
plt.scatter(x,y)
```

```python
x=np.random.randint(0,50,1000)
y=np.random.randint(0,10,1000)-x

np.corrcoef(x,y)
plt.scatter(x,y)
colors=np.random.rand(1000)
plt.scatter(x,y,c=colors,alpha=0.5)
plt.show()
```
